```python
import pandas as pd 
```

```python
pd.read_csv("50_Startups.csv") 
```

```python
 data = pd.read_csv("50_Startups.csv") 
```

```python
data
```

```python
data = pd.get_dummies(data,dtype = int,drop_first=True)
```

```python
data
```

```python
independent = data[["R&D Spend", "Administration" , "Marketing Spend","State_Florida","State_New York"]]	
```

```python
independent
```

```python
dependent = data[["Profit"]]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.40,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Regressor = LinearRegression() 
```

```python
Regressor
```

```python
Regressor.fit(x_train,y_train)
```

```python
weight = Regressor.coef_
```

```python
weight
```

```python
bias = Regressor.intercept_
```

```python
bias
```

```python
y_predict = Regressor.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score
```

```python
r_score = r2_score(y_test,y_predict)
```

```python
r_score
```
